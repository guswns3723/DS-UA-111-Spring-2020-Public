```python
import pandas as pd
import matplotlib.pyplot as plt     #we know we're going to use this since we're doing visualizations               
data = pd.read_csv('horses.csv')    #why stop when we've got such a great dataset
```

```python
data.head()                         #ah, there's our trusty dataset!
```

```python
data['temperament'].describe()        #confirm it's a continuous numeric variable (could just inspect type)
```

```python
plt.hist(data['temperament'])
plt.show()
```

```python
plt.hist(data['temperament'], bins=10)       #binning specifies how fine-grained our visualization is
plt.show()                                   #I usually start with default, then explore higher & lower bins
```

```python
plt.hist(data['temperament'], bins=200)         #what happens if you try bins = 4? 1? 1000?
plt.show()
```

```python
data['temperament'].mode()
```

```python
pd.set_option('display.max_rows', 1000)
data['temperament'].value_counts()
```

```python
data['temperament'].value_counts().plot(kind='bar')       #just because you can, doesn't mean you should!
plt.show()
```

```python
data['weight'].value_counts()           #we know this is actually a categorical variable -- let's confirm
```

```python
data['weight'].value_counts().plot(kind='bar')
plt.show()
```
