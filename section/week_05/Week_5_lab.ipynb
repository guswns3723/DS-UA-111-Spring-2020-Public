## View vs. Inplace

When we do something with a dateframe, pandas can do one of two things:

1. In can make the changes to the data frame, i.e., it does it inplace
2. Or, pandas can give us a view

A view is basically Pandas doing what we say, but not changing the data, just creating an alternative way to view the data -> without changing the data.

So if we want to sort and have it actually stay that way, we have to tell pandas to do it inplace.


```python
my_data.sort_values('price', ascending = False, inplace = True)
```

What if I only want to work with some of my data?

This is called a subset, or selecting a subset

Below we are selecting ONLY horses that are "Bay" colored


```python
my_data[my_data.color == "Bay"].head()
```

```python
my_data
```

Ok, we subsetted it, how does it work?

```python
my_data.color == "Bay"
```

How do we keep this subset? Well can assign it to a new variable name, like `bay_df`

If we assign it to the original variable name, we will lose all the horses.

```python
bay_df = my_data[my_data.color == "Bay"]
```

```python
bay_df.color.value_counts()
```

But, bay_df is just a view into my_data. If we change data in bay_df it may (depending on what we do) change `my_data`, which we don't want.

We want to copy our data and create our own little seperate place for it.

So, we'll use .copy()

```python
bay_df = my_data[my_data.color == "Bay"].copy()
```

```python
bay_df
```

Now we have the subset we want, and no matter what we do to it, our original data in `my_data` is safe.

# 4. <u> if statements </u>

Now what if we want python to figure things out for us?

For example, we if wanted python to go through some numbers and keep them only if they were prime.

The way we get python (or any programming language) to do these kind of "if this then that" sort of thing is called conditional statements, which is part of something larger called control flow.

let's see if statements in action.

```python
a = 14
b = 3    
```

```python
if a > b:
    
    print("a is greater than b")
    
if b > a:
    
    print("b is greater than a")
    
```

Notice the syntax, its just `if`, then anything that evaluates to a boolean, then parenthesis.

Python goes through and evaluates if the statement is True or False. 

If it is true, python goes through and does everything that is INDENTED under it.

```python
if a > b:
    
    print("a is greater than b")
    
if b > a:
    print("b is greater than a")
    
print("this always prints")
```

## Python is WHITESPACE delimited -> that means python uses empty space to figure out what the if statement is suppose to do if it is true. 

The other part of if statements is `else`. That means `if` the first statement isn't true, then do this. If the statement gets to end (meaning no if statement is true yet), it will ALWAYS excute the else statement.

```python
if b > a:
    print("b is greater than a")
else:
    print("a is greater than b")
```

Finally, there is `elif`, this is just "else if", meaning we can do a bunch of if statements together.

```python

if a == b:
    print("a equals b")
elif a > b:
    print("a is greater than b")
else:
    print("b is greater than a")
```

Importantly, python is going to stop once one of these statments is true!
