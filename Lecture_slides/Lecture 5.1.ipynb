# Lecture 5.1
Importing and and initial exploration of a dataset

```python
import pandas as pd                       #we need pandas for this!
data = pd.read_csv('Lecture5.1-horses.csv')          #this will only work if horses.csv is in the directory where I'm working
```

```python
data.head()      #experiment with putting different numbers in the () and see what happens; e.g. data.head(10)
```

```python
data.dtypes    #show types of each variable; but how to know this is the code? We tell you, or you SEARCH LIKE MAD
```

```python
data.shape      #one of my first go-tos when working with new data. This shows (rows, columns)
```

```python
data.describe()   #show common descriptive statistics for all numeric variables
```

```python
data['price'].describe()   #show common descriptive statistics for just the specified variable 'price'
```

```python
data['color'].describe()   #I actually can use describe for string variables; notice the different output
```

```python
data['color'].value_counts()     #I happen to love 'value_counts' for categorical variables
```

```python
data['name'].value_counts()  #not as useful for all variables, but I guess interesting there are some repeat names
```

```python
data['price'].mean()    #find the mean of just 'price'
```

# Tip:
To find out more about this syntax you can explore what [others have written](https://www.geeksforgeeks.org/python-pandas-dataframe-mean/) or [original documentation](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.mean.html).

```python
data.sort_values(by='price')     #let's find out the most expensive horse!! Sort the dataset by 'price'
```

```python
data.sort_values(by='price').head()  #eep, that was too many rows for my taste. We can add the head() syntax here
```

```python
data.sort_values(by='price', ascending=False).head() #we want to know the most expensive, this sorts high to low
```
